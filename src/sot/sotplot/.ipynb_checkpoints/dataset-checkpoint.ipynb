```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import io_refdata
```

```python
DSCOVR=np.array([[317.5,1],[325,2],[340,3],[388,3],[443,3],[551,3],[680,3],[687.75,0.8],[764,1],[779.5,2]])
tess=pd.read_csv("tess.csv",delimiter=",",names=["band","val"])
```

```python
refdata="/home/kawahara/exomap/sot/data/refdata"
cloud, cloud_ice, snow_fine, snow_granular, snow_med, soil, veg, ice\
, water, clear_sky=io_refdata.read_refdata(refdata)
```

```python
def getfx(i):
    ii=int(i)
    c=DSCOVR[ii,0]
    e=DSCOVR[ii,1]/2
    return [c-e,c+e,c+e,c-1]
```

```python
plt.plot(cloud[:,0]*1000,cloud[:,1],color="black",lw=1,ls="dashed",label="cloud")
plt.plot(soil[:,0]*1000,soil[:,1],color="black",lw=1,label="soil")
plt.plot(veg[:,0]*1000,veg[:,1],color="black",lw=1,ls="dotted",label="vegetation")
for i in range(0,9):
    plt.fill(getfx(i),[1,1,0,0],color="green",alpha=0.5)
plt.fill(getfx(9),[1,1,0,0],color="green",alpha=0.5,label="DSCOVR")
plt.plot(tess["band"]*1000,tess["val"],label="TESS")
plt.legend()
plt.xlim(300,1500)
plt.xlabel("wavelength (nm)")
plt.savefig("comp.png")
```
