```python
timg=img[::8,::8,:]
Nx, Ny, Nc = np.shape(timg)
N_data = 300
rand_now = module.random_generator(N_data, Nx, Ny)
Y, G=rand_now.make_colordata(timg,10) 

G=G.reshape(N_data,46*51)
G=cp.asarray(G)
Y=cp.asarray(Y)
```

```python
np.shape(Y)
```

```python
N=5
A0=np.random.rand(46*51,N)
X0=np.random.rand(N,np.shape(Y)[1])
```

```python
fig=plt.figure()

#timg=img[100:175,100:250]
#timg=img[100:175:2,100:250:2]

plt.imshow(timg)
```

```python
#no regularization
A=cp.asarray(A0)
X=cp.asarray(X0)
epsilon=1.e-6
omega=1.5
for i in range(0,1000):
    if np.mod(i,100)==0: print(i,cp.sum(Y - cp.dot(cp.dot(G,A),X))) #
    Wt = cp.dot(cp.dot(G.T,Y),X.T) #G^T Y X
    Wb = cp.dot(cp.dot(cp.dot(cp.dot(G.T,G),A),X),X.T) + epsilon # G^T G A X X^T
    #print(np.shape(Wt/Wb),np.shape(A))
    A = A*(Wt/Wb)**omega

    Wt = cp.dot(cp.dot(A.T,G.T),Y)
    Wb = cp.dot(cp.dot(cp.dot(cp.dot(A.T,G.T),G),A),X)+ epsilon 
    X = X*(Wt/Wb)**omega
```

```python
Iprex=cp.asnumpy(cp.dot(A,X).reshape(46,51,3))
fig=plt.figure()
plt.imshow(Iprex)
```

```python
A=cp.asarray(A0)
X=cp.asarray(X0)
epsilon=1.e-10
lam=1.e-4
for i in range(0,1000):
    if np.mod(i,100)==0: print(i,cp.sum(Y - cp.dot(cp.dot(G,A),X)))
    ATA = cp.dot(A.T,A)
    Wt = cp.dot(cp.dot(cp.dot(G.T,Y),X.T),ATA)+ epsilon
    Wb = cp.dot(cp.dot(cp.dot(cp.dot(cp.dot(G.T,G),A),X),X.T),ATA) + lam*cp.linalg.det(ATA)*A + epsilon
    #print(np.shape(Wt/Wb),np.shape(A))
    A = A*(Wt/Wb)
    A = cp.dot(A,cp.diag(1/cp.sum(A[:,:],axis=0)))
    Wt = cp.dot(cp.dot(A.T,G.T),Y)+ epsilon
    Wb = cp.dot(cp.dot(cp.dot(cp.dot(A.T,G.T),G),A),X)+ epsilon 
    X = X*(Wt/Wb)
```

```python
Ipre=cp.asnumpy(cp.dot(A,X).reshape(46,51,3))
```

```python
fig=plt.figure()

#timg=img[100:175,100:250]
#timg=img[100:175:2,100:250:2]
Ax=cp.asnumpy(A)
c=plt.imshow(Ax[:,2].reshape(46,51))
plt.colorbar(c)
```
